```python
ds = load_dataset("DBD-research-group/BirdSet", region, trust_remote_code=True)
```

```python
print(ds[split][0]['audio'])
```

```python
for item_idx, example in enumerate(ds[split]):
  stripped_path = example['audio']['path'].split("/")[-1]
  
  filename_sam = os.path.join(output_dir, f"sam_{stripped_path}")

  csv_path_sam = os.path.join(output_dir, METADATA_FILE)

  with open(csv_path_sam, mode='a', newline='', encoding='utf-8') as f:
      writer = csv.writer(f)

      row = [f"sam_{stripped_path}"] + [example[col] for col in example if col not in ["audio", "filepath"]]

      writer.writerow(row) 
```
